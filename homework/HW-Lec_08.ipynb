```python
import matplotlib.pyplot as plt
import numpy as np
left, right = 0, 3
grain = 100
order = 10

plt.xlabel('x')
plt.ylabel('y')
plt.xlim([left, right])
plt.legend(fontsize=12)



#f = lambda x: x**2 / (x+1)**(1/2)
#f = lambda x: x**3 / (x**2 + 1)
f = lambda x: np.sin(x)



x = np.linspace(left, right, grain)
y = f(x) + np.random.rand(grain) - 0.5

def getError(valArray):
    j = 0
    total_error = 0
    for j in range(valArray.size):
        #rint(j)
        #rint(f(3*j/100))
        total_error += (valArray[j] - f(3*j/100))
        #print(abs(valArray[i] - f(3j/100)))
    return abs(total_error)

i = 1
minError = 200 #It'll never reach this high and being lazy like this is a valid strategy, plus you can't stop me
best_order = 0
for i in range(1, order+1) :
    plt.figure(figsize=[10, 6])
    fit = np.polyfit(x, y, i, full=True)
    y_p = np.polyval(fit[0], x)
    plt.plot(x,f(x), label='correct')
    plt.plot(x,y, '.', label='mess')
    plt.plot(x,y_p, label='guess ' + str(i))
    #plt.title('fitted coefficients' + str(fit[0]) + '\n residual: ' + str(fit[1]))
    plt.show()
    error = getError(y_p)
    if error < minError:
        minError = error
        best_order = i
    print(error)
    print(i)
#plt.text(2, 2, fit)
print("Best order was " + str(best_order) + ", with error of " + str(minError) + ".")
#print(np.polyval(fit[0], x))
```
